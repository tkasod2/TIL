**1. 지니 불순도(gini inpurity) 지표를 계산하는 함수를 구현해 보세요**

![gini impurity](https://miro.medium.com/max/490/1*X7CrquTnWeYDSXHrHBvm-A.png)

* 입력으로 클래스 확률 분포를 받습니다.
* 확률 분포는 숫자 리스트로 표현되며, 숫자는 모두 0보다 같거나 크고 합하면 1이 되는 소수입니다.
* 출력으로 주어진 공식에 따라 계산된 지니 불순도를 값을 반환합니다.
* 라이브러리(내장/외장)를 사용하지 말고 파이썬의 기능만으로 구현해보세요.

```python
def gini_impurity(p_list):
  # =================================================================================
  # 여기에 코드를 입력하세요
  # 입력 타입은 숫자 리스트(list of floats)입니다.
  # 출력 타입은 숫자(float) 입니다.
  gini = 0
  for pi in p_list:
    gini += pi*(1-pi)

  return gini
  # pass
  # =================================================================================
```

```python
# 테스트 코드입니다.
# 함수가 제대로 구현되었는지 평가를 위한 코드이니 수정하지 마세요.
# 실행했을 때 아무 에러도 뜨지 않으면 성공입니다.
# 프로그래밍 언어의 특성 상 딱 떨어지는 숫자로 값이 나오지 않을 수 있기 때문에 정답과 차이가 충분히 작으면 정답으로 인정합니다.

epsilon = 0.001 # small number
assert abs(gini_impurity([0, 1]) - 0) < epsilon
assert abs(gini_impurity([0, 0, 1]) - 0) < epsilon
assert abs(gini_impurity([0.5, 0.5]) - 0.5) < epsilon
assert abs(gini_impurity([0.3, 0.3, 0.4]) - 0.66) < epsilon
```

**2. 의사 결정 나무(decision tree)를 이용해 이진 분류 모델 만들기**

* 이탈 예측 모델에서 배웠던 의사 결정 나무를 이용합니다
* 이탈 예측과 같은 이진 분류 문제인 유방암 분류 데이터셋에서 의사 결정 나무를 학습시켜 봅니다
* 테스트셋에서 정확도가 85% 이상이면 통과입니다

```python
# 데이터셋 준비
# 이 코드는 수정하지 않습니다
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn import metrics
import pandas as pd

cancer_data = load_breast_cancer()
df = pd.DataFrame(cancer_data.data, columns=cancer_data.feature_names)

# 데이터셋 살펴보기
df
```

```python
# 학습/테스트 데이터셋 나누기
# 이 코드는 수정하지 않습니다
df_train, df_test, y_train, y_test = train_test_split(df, cancer_data.target, random_state=0)
```

```python
# 의사결정 나무 모델을 불러옵니다
from sklearn.tree import DecisionTreeClassifier
```

```python
DecisionTreeClassifier?
```

```python
# =================================================================================
# 여기에 코드를 입력하세요
# 모델 초기화 및 학습을 진행합니다
# 학습된 모델은 "tree_model"이란 변수에 저장하세요

clf = DecisionTreeClassifier()
tree_model = clf.fit(df_train, y_train)

# =================================================================================
```

```python
# 모델 평가하기 (정확도)
# 이 코드는 수정하지 마세요
# tree_model의 정확도가 85% 이상이면 통과입니다
prediction_test = tree_model.predict(df_test)
print(metrics.accuracy_score(y_test, prediction_test))
```

**3. 랜덤 포레스트(Random Forest)를 이용해 트리 앙상블 모델 만들기**

* 트리 앙상블 강의에서 설명했던 랜덤 포레스트를 학습시켜 봅니다
* 위에서 실습했던 데이터셋을 그대로 활용하여 이진 분류 문제를 풉니다

```python
# 랜덤 포레스트 모델을 불러옵니다
from sklearn.ensemble import RandomForestClassifier
```

```python
# =================================================================================
# 여기에 코드를 입력하세요
# 모델 초기화 및 학습을 진행합니다
# 학습된 모델은 "rf_model"이란 변수에 저장하세요
clf_rf = RandomForestClassifier()
rf_model = clf_rf.fit(df_train, y_train)

# =================================================================================
```

```python
# 모델 평가하기 (정확도)
# 이 코드는 수정하지 마세요
# rf_model의 정확도가 95% 이상이면 통과입니다
prediction_test = rf_model.predict(df_test)
print(metrics.accuracy_score(y_test, prediction_test))
```
